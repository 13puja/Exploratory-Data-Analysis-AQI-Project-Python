```python
#1.What is the total number of records in the AQI dataset?
station_records=station_hour.shape[0]
print("total number of records:",station_records)
```

```python
city_records=city_hour.shape[0]
print("total number of record :", city_records)
```

```python
hour_data['AQI']=pd.to_numeric(hour_data['AQI'],errors='coerce')
```

```python
hour_data=hour_data.dropna(subset=['AQI'])
```

```python
#2.What is the average AQI value for each city?
avg_city_aqi=hour_data.groupby("City")["AQI"].mean(numeric_only=True)
print(avg_city_aqi)
```

```python
plt.figure()
avg_city_aqi.plot(kind='bar')
plt.xlabel('city')
plt.ylabel('average AQI')
plt.show()
```

```python
#3.What is the maximum and minimum AQI recorded in the dataset?
max_aqi=station_hour['AQI'].max()
print("maximum AQI is:",max_aqi)
min_aqi=station_hour['AQI'].min()
print("minimum AQI is:",min_aqi)
```

```python
min_aqi=city_hour['AQI'].min()
print("minimum AQI is:",min_aqi)
```

```python
#4.Which city has the highest average AQI?
avg_city = hour_data.groupby('City')['AQI'].mean(numeric_only=True)
print("highest avg aqi city:",avg_city.idxmax())
avg_city.sort_values(ascending=False).plot(kind='bar')
plt.show()
```

```python
#5.Which city has the lowest average AQI?
avg_city = hour_data.groupby('City')['AQI'].mean(numeric_only=True)
print("lowest avg aqi city:",avg_city.idxmin())
avg_city.sort_values(ascending=True).plot(kind='bar')
plt.show()
```

```python
day_data['Date']=pd.to_datetime(day_data['Date'])
day_data['YEAR']=day_data['Date'].dt.year
day_data['Month']=day_data['Date'].dt.month
day_data['month_name']=day_data['Date'].dt.month_name()
```

```python
day_data['AQI']=pd.to_numeric(day_data['AQI'],errors='coerce')
```

```python
day_data=day_data.dropna(subset=['AQI'])
```

```python
#8.Which year has the highest recorded AQI?
day_data['Date']=pd.to_datetime(day_data['Date'])
day_data['YEAR']=day_data['Date'].dt.year
year_highest_aqi=day_data.loc[day_data['AQI'].idxmax(),'YEAR']
year_highest_aqi
```

```python
#9.Which year has the lowest recorded AQI?
day_data['Date']=pd.to_datetime(day_data['Date'])
day_data['YEAR']=day_data['Date'].dt.year
year_minimum_aqi=day_data.loc[day_data['AQI'].idxmin(),'YEAR']
year_minimum_aqi
```

```python
#10.What is the average AQI for each month across all years?
monthly_avg_aqi=day_data.groupby('month_name')['AQI'].mean().sort_values()
monthly_avg_aqi
monthly_avg_aqi.plot(kind='bar', figsize=(10,5),title='avg aqi by month')
plt.show()
```

```python
#11.What is the total number of records for each air quality category (e.g., Good, Moderate, Unhealthy)?
aqi_category_count=city_day['AQI_Bucket'].value_counts()
aqi_category_count
aqi_category_count.plot(kind='bar', figsize=(10,5),title='aqi_category_count')
plt.show()
```

```python
#12.What is the percentage of days that have a 'Good' AQI across all cities?
good_days=(city_day['AQI_Bucket']=='Good').sum()
total_days=city_day.shape[0]
percentage_good=(good_days/total_days)*100
percentage_good
```

```python
city_day['Date']=pd.to_datetime(city_day['Date'])
city_day['YEAR']=city_day['Date'].dt.year
city_day['Month']=city_day['Date'].dt.month
city_day['month_name']=city_day['Date'].dt.month_name()
```

```python
city_day['AQI']=pd.to_numeric(city_day['AQI'],errors='coerce')
```

```python
city_day=city_day.dropna(subset=['AQI'])
```

```python
#13.How does the AQI change on a yearly basis for each city?
yearly_city_aqi=day_data.groupby(['City','YEAR'])['AQI'].mean(numeric_only=True).reset_index()
yearly_city_aqi
```

```python
#14.What are the top 5 cities with the most days of "Unhealthy" AQI?
unhealthy_cities=(city_day[city_day['AQI_Bucket']=='Unhealthy'].
                  groupby('City').size().sort_values(ascending=False).head(5))
unhealthy_cities
```

```python
#15.Which months tend to have the worst air quality on average?
worst_months = city_day.groupby('month_name')['AQI'].mean().sort_values(ascending=False)
print(worst_months)
```

```python
#16.What is the percentage of missing AQI data per city?
missing_aqi=city_day.groupby('City')['AQI'].apply(lambda x:x.isna().mean()*100)
missing_aqi
```

```python
#17.Find the cities where AQI has shown the most improvement (lower AQI) from the previous year.
aqi_yearly=city_day.groupby(['City','YEAR'])['AQI'].mean().reset_index()
aqi_pivot=aqi_yearly.pivot(index='YEAR',columns='City', values='AQI')

aqi_improvement=aqi_pivot.diff().mean().sort_values()
aqi_improvement.head()
```

```python
#28.For each air quality level, what is the average number of days that fall into each category in a year?
avg_days_category=(city_day.groupby(['YEAR','AQI_Bucket']).size().groupby('AQI_Bucket').mean())
avg_days_category
```

```python
#19.What is the distribution of AQI values for each air quality level (e.g., Good, Moderate, Unhealthy)?
plt.figure(figsize=(10,6))
sns.boxplot(data=city_day, x='AQI_Bucket', y='AQI', showfliers=False)
plt.title('distribution of AQI values by Air quality category')
plt.xticks(rotation=30)
plt.show()
```
